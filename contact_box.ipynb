```python
import tkinter as tk

from tkinter import *

path="database.txt"

def savefile(line):
        file = open(path,'a',encoding='utf-8')
        file.writelines(line)
        file.writelines("\n")
        file.close()

def ReadFile():
    arrcontacts=[]
    file=open(path,'r', encoding='utf-8')
    for line in file:
        data=line.strip()
        arr=data.split(";")
        arrcontacts.append(arr)
    file.close()
    return arrcontacts

def addcontact():   
    line = stringName.get()+ ";" +stringEmail.get() + ";" + stringPhonenumber.get() +";"+ stringNote.get() 
    savefile(line)
    stringName.set(" ")
    stringEmail.set(" ")
    stringPhonenumber.set(" ")
    stringNote.set(" ")
    showcontacts()
    
def deletecontact2():
    with open("database.txt", "r") as f: lines = f.readlines()   
    with open("database.txt", "w") as f:
        for line in lines:
            if line.strip("\n") != ANCHOR:
                f.write(line)
    listbox.delete(ANCHOR)

def showcontacts():
    arrcontacts= ReadFile()
    listbox.delete(0,END)
    for item in arrcontacts:
        listbox.insert(END,item) 
             
def findcontact():
    arrcontacts = ReadFile()
    ma= stringName.get()
    found=False
    for contact in arrcontacts:
        if contact[0]== ma:
            found= True
            break
    if found:
        pass
    else:
        pass

def deletecontact():
    listbox.delete(0, tk.END)
    
    
def deleteselected():
    listbox.delete(ANCHOR)
    
    
root = Tk()

stringName=StringVar()
stringEmail=StringVar()
stringPhonenumber=StringVar()
stringNote=StringVar()

# lam man hinh voi do dai va co gian duoc
root.title("friends")
root.resizable(height=True, width=True)
root.minsize(height=370,width=360)



Label(root, text="My contacts", fg='blue',font=("Helvetica", 16)).grid(row=0,columnspan=2)


listbox = Listbox ( root, width=50, fg = 'blue', justify = CENTER)

listbox.grid(row=1,columnspan=2)

showcontacts()

#lam cac o nhap thong tin name. email, phonenumber va note

Label(root, text="Name:").grid(row=2,column=0)
e1 = Entry(root,width = 30,textvariable=stringName).grid(row=2,column=1)

Label(root, text="Email:").grid(row=3,column=0)
e2 = Entry(root,width = 30,textvariable=stringEmail).grid(row=3,column=1)

Label(root, text="Phone Number:").grid(row=4,column=0)
e3 = Entry(root,width = 30,textvariable=stringPhonenumber).grid(row=4,column=1)

Label(root, text="Note:").grid(row=5,column=0)
e4 = Entry(root,width = 30,textvariable=stringNote).grid(row=5,column=1)

Label(root, text="Search Name:").grid(row=7,column=0)
Entry(root,width = 30).grid(row=7,column=1)

#lam cai nut add, search, arrange, exit

frameButton=Frame(root)


Button(frameButton,text="Add", command=addcontact).pack(side=LEFT)

Button(frameButton,text="Search", command=findcontact).pack(side=LEFT)

Button(frameButton,text="Delete All",command=deletecontact).pack(side=LEFT)

Button(frameButton,text="Delete Select",command=deleteselected).pack(side=LEFT)

Button(frameButton,text="Delete contact2",command=deletecontact2).pack(side=LEFT)

Button(frameButton,text="Exit", command=root.quit).pack(side=LEFT)

frameButton.grid(row=8,column=1, pady=5)
```

```python
root.mainloop()
```
